```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alphas = np.linspace(1.5, 2, 1000)
plt.plot(alphas, (2/alphas-1)*2**(2*alphas+1)*(2/alphas - 1/2) - 1)
```

```python
plt.plot(alphas, (2/alphas -1) * (2/alphas - 1/2) * 2**(2*alphas+1) - 1)
```

```python
for i, val_i in enumerate((2/alphas -1) * (2/alphas - 1/2) * 2**(2*alphas+1) - 1):
    if val_i < 0:
        print(alphas[i-1], val_i)
        break
```
