```python
def f(alpha):
    return 2**(alpha + 1/2)* (2/alpha - 1) * (2/alpha - 1/2) - 2**(-alpha - 1/2)


import numpy as np
x = np.linspace(1, 2, 100)

import matplotlib.pyplot as plt

plt.plot(x, f(x))
plt.plot(x, np.zeros(100))

```

```python
from scipy.optimize import newton
```

```python
a = newton(f, 1.5)
```
