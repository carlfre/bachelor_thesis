```python
def ub(alpha):
    return 3**(-alpha - 1)/2 + 3**(-alpha)/alpha + (alpha+1)/12 * 3**(-alpha - 2)

# import riemann zeta
from scipy.special import zeta
def true(alpha):
    return zeta(alpha+1) - 1 - 2**(-alpha - 1)
```

```python
import numpy as np
alphas = np.linspace(0.1, 2, 100)

import matplotlib.pyplot as plt
plt.plot(alphas, ub(alphas))
plt.plot(alphas, true(alphas))
plt.legend(["Upper bound", "True value"])
plt.show()
```

```python
plt.plot(ub(alphas) - true(alphas))
```

```python
def lhs(alpha):
    return ub(alpha) * (1 + 2**(2*alpha)*(2/alpha - 1))**2 / (2**(2*alpha +1) *(2/alpha - 1)*(2/alpha - 1/2) -1)


def rhs(alpha):
    m = 3
    return  -ub(1) + (2/alpha - 1) *m**alpha * (2/alpha  - m**(1/2)*3**(alpha - 1/2)*ub(alpha))
```

```python
alphas = np.linspace(0.5, 2, 100)
plt.plot(alphas, lhs(alphas))
plt.plot(alphas, rhs(alphas))
plt.legend(["LHS", "RHS"])
plt.show()

```
