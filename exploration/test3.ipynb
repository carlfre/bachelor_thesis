```python
import numpy as np

# import zeta function
from scipy.special import zeta

import matplotlib.pyplot as plt
```

```python
def g(alpha):
    nom = (1 + 2**(2*alpha) *(2/alpha - 1))**2
    denom = 2**(2*alpha + 1) * (2/alpha - 1/2) * (2/alpha - 1) - 1
    return 1 + nom/denom


def h(alpha):
    return (2/alpha - 1) * (1 + 2**(alpha - 1)) + (2**(-alpha - 1) + 1)*g(alpha)

# def zeta_ub(alpha):
#     am = -alpha
#     ub = 1 + 2**am + 3**am + 4**am


def A(alpha):
    return 2693/1920 - np.sqrt(2)/24 - 2/alpha

```

```python
alphas = np.linspace(1.49, 1.53, 100)

def lhs(alpha):
    return (1 + 2**(-alpha - 1) + 3**(-alpha - 1) + 4**(-alpha - 1) + 4**(-alpha)* (1/alpha - 1/8 + (alpha + 1)/192)) * g(alpha)

def rhs(alpha):
    return h(alpha) + (2/alpha - 1) * (- alpha/6 * 3**(alpha - 2) + (3-alpha)*(5-alpha)*(6-alpha)*(8+alpha)/(720*alpha))



alpha = alphas
# def lhs(alpha):
#     return zeta(alpha+1)*g(alpha)

# def rhs2(alpha):
#     return h(alpha) + A(alpha)

plt.plot(alphas, lhs(alphas), label='lhs')
# plt.plot(alphas, zeta(alphas+1)*g(alphas))
plt.plot(alphas, rhs(alphas), label='rhs')
# plt.plot(alphas, rhs2(alphas), label='rhs2')
plt.legend()
plt.show()
```
